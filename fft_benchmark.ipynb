```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams.update({"font.size": 14})
```

```python
from timeit import timeit
```

```python
data = np.arange(2048)
```

```python
N = 20_000
lengths = np.arange(1010, 1039)
times = np.fromiter(
    map(
        lambda l: timeit(f"np.fft.fft(data[:{l}])", number=N, globals=globals()),
        lengths,
    ),
    float,
)
```

```python
normalized = times / (lengths * N) * 1e9
plt.plot(lengths, normalized, "o-")
plt.ylabel("Time taken [ns/sample]")
plt.xlabel("FFT length [samples]")
plt.title("FFT benchmark")
plt.savefig("fft_benchmark.pdf")
plt.show()
```
